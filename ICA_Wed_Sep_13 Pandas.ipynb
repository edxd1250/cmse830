```python
def bill_less_avg(val):
    """
    turns green if top 90% and red if bottom 10% of average
    """
    avg = val.mean()
    if val < avg : color = 'red' 
    else : color = 'black'
    return 'color: %s' % color

df_peng.style.apply(bill_less_avg, subset= notin(df_peng, ['bill_length_mm','bill_depth_mm']))



```
